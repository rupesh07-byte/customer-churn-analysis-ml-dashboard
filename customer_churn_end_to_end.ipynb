## Prepare Data for Model

```python
df_model=pd.get_dummies(df, drop_first=True)
```

### 🔥14. Split Data

```python
from sklearn.model_selection import train_test_split
x= df_model.drop("Churn",axis=1)
y=df_model["Churn"]

X_train, X_test, y_train, y_test =train_test_split(x, y, test_size=0.2, random_state=42)
```

### 🔥 15. Train Model

```python
from sklearn.linear_model import LogisticRegression

model= LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

### 🔥16. Checked Accuracy

```python
from sklearn.metrics import accuracy_score
y_pred= model.predict(X_test)
accuracy_score(y_test, y_pred)
```

### 17. Check Model Performance

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

Meaning:<br>
915 → Correctly predicted NON-churn ✅<br>
118 → Predicted churn but actually stayed ❌<br>
181 → Missed churn customers ❌ (IMPORTANT)<br>
193 → Correctly predicted churn ✅<br>

### 🎯 Insight

💡 The model achieves ~79% accuracy but has low recall (52%) for churn customers, meaning it misses a significant portion of at-risk customers.

🚀 Business Recommendation 

👉 Improve churn detection by:<br>

Using class balancing techniques<br>
Trying advanced models (Random Forest, XGBoost)<br>
Focusing on improving recall over accuracy

### In churn prediction, recall is more important than accuracy because identifying at-risk customers is critical for retention strategies.
